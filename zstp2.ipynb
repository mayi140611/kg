```python
from pymongo import MongoClient
```

```python
client = MongoClient()
db = client['mkb']
collection = db['medicineChdetail']
```

```python
'''
图数据库的建立
'''
#Graph()实例化
from py2neo import Graph
graph = Graph("http://localhost:7474",user='neo4j',password='123')
from py2neo import Node, Relationship,size ,order
```

```python
import re
re.sub(r'\s+','','别    称')
```

```python
count = 0
for c in collection.find({}):
#     print(c)
    if count < 50 : 
        count += 1
        continue
    if count > 500 : break
    for k,v in c.items():
        k = k.replace('dd','')
        k = re.sub(r'\s+','',k)
        if k == 'title' or k=='_id': 
            continue
        s = 'merge (n:中医药 {name:"'+c['title']+'"}) merge (n1:'+k+'{name:"'+v+'"}) create (n)-[:'+k+']->(n1)'
        print(s)
        graph.run(s)
    count += 1
```
