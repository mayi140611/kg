```python
import pymongo
from pymongo import MongoClient

client = MongoClient('localhost', 27017)

#Getting a Database
db = client['baidubaike']
```

## 处理疾病诊断分类
* 疾病实体含有的可能属性并集
* 疾病实体
* 科室实体
* 症状实体
* 发病部位
* 多发群体

```python
c = db['symptomsdetail']
```

```python
#提取所有疾病实体的属性
s = set()
for t in c.find(no_cursor_timeout=True):
#     print(type(t))
#     print(set(t.keys()))
    s = s.union(set(t.keys()))
#     break

len(s)#826
```

```python
list(s)
```

```python
#提取所有疾病实体
diseases = set()
for t in c.find(no_cursor_timeout=True):
    diseases.add(t['title'])
len(diseases)#6771，和爬取网页数相同，证明没有重复
```

```python
diseases
```

```python
#提取就诊科室
department = set()
for t in c.find(no_cursor_timeout=True):
    if 'dd就诊科室' in list(t.keys()):  
        t['dd就诊科室'] = t['dd就诊科室'].replace('、','，').replace('；','，')
        tt = t['dd就诊科室'].split('，')
        department = department.union(list(map(lambda x:x.strip(),tt)))
                         
len(department)#163
```

```python
department
```

```python
#提取症状
symptom = set()
for t in c.find(no_cursor_timeout=True):
    if 'dd常见症状' in list(t.keys()):  
        t['dd常见症状'] = t['dd常见症状'].replace('、','，').replace('：','，').replace('；','，').replace('。','，').replace('或','，').replace('和','，').replace('等','')
        tt = t['dd常见症状'].split('，')
        symptom = symptom.union([i for i in list(map(lambda x:x.strip(),tt)) if i!='' and 1<len(i)<11])
                         
len(symptom)#12366
```

```python
symptom
```

```python
#提取发病部位
position = set()
for t in c.find(no_cursor_timeout=True):
    if 'dd常见发病部位' in list(t.keys()):  
        t['dd常见发病部位'] = t['dd常见发病部位'].replace('、','，').replace('：','，').replace('；','，').replace('。','，').replace('或','，').replace('和','，').replace('等','')
        tt = t['dd常见发病部位'].split('，')
        position = position.union([i for i in list(map(lambda x:x.strip(),tt)) if i!='' and 1<len(i)<11])
                         
len(position)#1405
```

```python
position
```

```python
#提取多发群体
group = set()
for t in c.find(no_cursor_timeout=True):
    if 'dd多发群体' in list(t.keys()):  
        t['dd多发群体'] = t['dd多发群体'].replace('、','，').replace('：','，').replace('；','，').replace('。','，').replace('或','，').replace('和','，').replace('等','')
        tt = t['dd多发群体'].split('，')
        group = group.union([i for i in list(map(lambda x:x.strip(),tt)) if i!='' and 1<len(i)<11])
                         
len(group)#12366
```

```python
group
```

```python
import pickle
with open('diseaseattr.pkl', 'wb') as f:
    pickle.dump(s, f)
    
with open('diseases.pkl', 'wb') as f:
    pickle.dump(diseases, f)
    
with open('symptoms.pkl', 'wb') as f:
    pickle.dump(symptom, f)
    
with open('positions.pkl', 'wb') as f:
    pickle.dump(position, f)
    
with open('departments.pkl', 'wb') as f:
    pickle.dump(department, f)
    
with open('groups.pkl', 'wb') as f:
    pickle.dump(group, f)
```
